1. Descargar el instalador y ejecutable de ANACONDA en el sitio oficial [Descargar Anaconda](https://www.anaconda.com/products/individual)


2. Ejecutar el instalador de ANACONDA, recomendacion ejecutar como administrador 


3. Al elegir dnde instalar el software en la pestaa "Seleccionar destino", seleccione Instalar solo para m "Just me (recommended)" . De lo contrario, necesitar acceso de administrador para realizar actualizaciones y acceder a algunos elementos desde la lnea de comandos. Adems, es una buena prctica de seguridad instalar solo localmente.
 


4. Seleccione la ubicacin de instalacin para Anaconda ( puede seleccionar el valor predeterminado del disco local C u otra ubicacin personalizada, debe ser ubicacin fija). Haga clic en Siguiente cuando est listo para continuar. 


5. Haga clic en la casilla de verificacin para registrar Anaconda como su Python predeterminado. No agregue Anaconda a su variable PATH. 


6. Haga click en instalar para iniciar la instalacin de ANACONDA. 


7. Una vez finalizada la instalacin, no elija instalar PyCharm IDE. Usaremos Jupyter Lab en este curso y PyCharm no ser compatible. 


8. Verifique que la instalacin sea exitosa iniciando Anaconda Navigator desde su men de inicio.
